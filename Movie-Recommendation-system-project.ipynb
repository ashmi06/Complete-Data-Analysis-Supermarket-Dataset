```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
#A Counter is a dict subclass for counting hashable objects. It is a collection where elements are stored 
#as dictionary keys and their counts are stored as dictionary values.
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
#Concatenating the columns
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']
```

```python
new_df= movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head(3)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
#checking for a particular column
new_df['tags'][0]
```

```python
#Coverting everything to lower case
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector.shape
```

```python
 #most frequent words that occured
cv.get_feature_names()
```

```python
len(cv.get_feature_names())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
