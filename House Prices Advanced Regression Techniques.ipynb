# Temporal Variables(Eg:Datetime Variables)

From the df we have 4 year variables. we have extract information from the datetime variables like no of years or no of days. One example in this specific scenario can be difference in years between the years the house was built and the year the house was sold. we will be performing this analysis in the feature Engineering 

```python
#List of variable that contains year information
year_features=[features for features in numerical_variable if 'Yr' in features or 'Year' in features]
year_features
```

```python
for feature in year_features:    #after writing this line for feature in year_features: i am saying that take feature from above list year_features
    print(feature, df[feature].unique())   #so df[features ] will be ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt', 'YrSold']
                                           #bcoz i am writing this line under (:)
```

```python
#Lets analyze the Temporal Datetime Variables
#We will check whether there is a relation between year the house is sold and sales price
#this groupby is telling me that take Yrsold column and salesPrice median for every YrSold row
df.groupby('YrSold')['SalePrice'].median().plot()    #taking median of all the (2008 year)rows,2007 year row and so on
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House price vs YearSold')

#Here we can see that as the year increases sales price decreases
```

```python
year_features
```

```python
#Here we will compare the difference between all the year features with salesprice
for features in year_features:
    if features!= 'YrSold':
        data=df.copy()
        data[features]=data['YrSold']-data[features]
        plt.scatter(data[features],data['SalePrice'])
        plt.xlabel(features)
        plt.ylabel('SalesPrice')
        plt.show()
```

```python
#<25 is giving me unique 25 valuess in row in numerica_variable feature
discrete_feature=[feature for feature in numerical_variable if len(df[feature].unique())<25 and feature not in year_features+['Id']]
print("Discrete Variables Count: {}".format(len(discrete_feature)))
```

```python
continuous_feature=[feature for feature in numerical_variable if feature not in discrete_feature+ year_features+['Id']]
print("Continuous Variable Count: {}".format(len(continuous_feature)))
```

```python
#Lets analyse the continuous values by creating histograms to understand the data
for feature in continuous_feature:
    data=df.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()    
```

```python
#We will be using logarithmic transformation
for feature in continuous_feature:
    data=df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Saleprice')
        plt.title(feature)
        plt.show()
```
